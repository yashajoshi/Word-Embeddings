```python
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
df = pd.read_csv('nll_results_minus_two.csv', sep='\t')
df = df[(df['0.1']!=0) & (df['0']!=0)]
#ax = df.plot('0.1','0', title='Problem 5. [-2, -1, 1, 2]', legend=False)
#ax.set_xlabel("IterationNumber")
#ax.set_ylabel("Cumulative Negative Log-likelihood")
sns.regplot(df['0.1'],df['0'], fit_reg=False)
```
